# Exploratory Data Analysis

## Data Collection

### Load Data


```python
import pandas as pd

df = pd.read_csv('../data/raw/review_tiktok_dataset.csv')   
```

### Data Summary

```python

df.info()
```

### Pengecekan Missing Value

```python
df.isna().sum()
```

### Distribusi Data Berdasarkan Score Review

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="score", hue="score", palette="viridis")

plt.title("Distribusi Skor Review")
plt.xlabel("Skor")
plt.ylabel("Jumlah")
plt.grid(linestyle='--', linewidth=0.5, color='gray', axis='y')
plt.tight_layout()
plt.show()
```

### Melihat Kata yang Paling Sering Muncul

```python
from wordcloud import WordCloud

text = " ".join(review for review in df['content'].astype(str))

wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color='black',
    max_words=100,
    colormap='viridis',
    contour_color='steelblue',
    contour_width=1
).generate(text)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud dari Kolom Review", fontsize=22, weight='bold')
plt.tight_layout(pad=0)
plt.show()
```
