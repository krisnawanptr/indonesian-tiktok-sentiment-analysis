```python
import pandas as pd

df = pd.read_csv('../data/preprocessed/review_tiktok_dataset_preprocessed.csv')
df.head()
```

```python
from transformers import pipeline
from tqdm import tqdm
import pandas as pd

tqdm.pandas()

sentiment_pipeline = pipeline("sentiment-analysis", model="Aardiiiiy/indobertweet-base-Indonesian-sentiment-analysis")

def get_sentiment_info(review):
    if isinstance(review, str) and review.strip():
        result = sentiment_pipeline(review)[0]
        return result['label'], result['score']  
    return 'neutral', 0.0


df[['sentimen', 'confident_score']] = df['content'].progress_apply(lambda x: pd.Series(get_sentiment_info(x))
)
```
